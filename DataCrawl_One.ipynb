```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import re
import numpy as np
```

```python
def dataout(response):
    soup = BeautifulSoup(response.content, 'html.parser')
    values = soup.find_all('div', class_='s-data-columns__value')
    if len(values) == 0:
        print('No values found')
    # Loop over each product and extract the data
    signalTrades = values[0].text 


    # Get total number of profit trades
    s = values[1].text
    number_regex = r"([\d\s]+)\s*\(.*\)"
    match = re.search(number_regex, s)
    if match:
        number = match.group(1)
        # Remove spaces and normalize the number
        normalized_number = float(number.replace(" ", "").replace(",", ""))
    signalProfitTrades = normalized_number

    # Get total number of loss trades

    s = values[2].text
    number_regex = r"([\d\s]+)\s*\(.*\)"
    match = re.search(number_regex, s)
    if match:
        number = match.group(1)
        # Remove spaces and normalize the number
        normalized_number = float(number.replace(" ", "").replace(",", ""))
    signalLossTrades = normalized_number

    signalBestTrade = values[3].find('i').text
    signalWorstTrade = values[4].find('i').text

    signalGrossProfit = values[5].find('i').text
    signalGrossLoss = values[6].find('i').text



    signalSharpeRatio = values[9].text

    signalRecoveryFactor = values[15].text


    signalProfitFactor = values[18].text

    signalExpectedPayoff = values[19].find('i').text

    # Find the <svg> element

    if "Algo trading" in values[25].parent.text:
        pattern = r'%'
        signalAlgoTrading = re.sub(pattern, '', values[25].text)
    else :
        pattern = r'%'
        signalAlgoTrading = re.sub(pattern, '', values[26].text)

    values2 = soup.find_all('div', class_='s-list-info__value')

    signalMonthlyGrowth = re.sub(pattern, '', values[24].text)

    signalGrowth = re.sub(pattern, '', values2[0].text) #percentage
    with open('signal_detail_one.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['19191',signalGrowth,signalTrades,signalProfitTrades,signalLossTrades,signalBestTrade,signalWorstTrade,signalGrossProfit,signalGrossLoss,signalSharpeRatio,signalRecoveryFactor,signalProfitFactor,signalExpectedPayoff,signalMonthlyGrowth,signalAlgoTrading])
```

```python

base_url_singleID = "https://stackoverflow.com/questions/47260995/visual-studio-code-tabs-editors-of-the-same-project-in-multiple-screens"
with open('signal_detail_one.csv', 'w', newline='') as signal_detailsFile:
  writer = csv.writer(signal_detailsFile)
  writer.writerow(['Signal Id','SignalGrowth','Trades','ProfitTrades','LossTrades','BestTrades','WorstTrades','GrossProfit','GrossLoss','SharpeRatio','RecoveryFactor','ProfitFactor','ExpectedPayoff','MonthlyGrowth','AlgoTrading'])




#proxy information
for index, row in proxies_list.iterrows():
    proxy_info = {
        'http': str(row['ip']) + ':' + str(row['port']),
        'https': str(row['ip']) + ':' + str(row['port'])
    }

    try:
        response = requests.get(base_url_singleID, proxies= proxy_info, timeout= 2)
        if response.status_code == 200:
            
            print ("succeed connection: ", proxy_info)
        else:
            print(f"Proxy {proxy_info} connection failed. Status code: {response.status_code}")
    except requests.RequestException as e:
            print(f"An error occurred while connecting to proxy {proxy_info}. Error: {str(e)}")



```
