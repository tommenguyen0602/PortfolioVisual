```python
import pandas as pd
import random 
import matplotlib.pyplot as plt
import pandas as pd
signals = pd.read_csv("Cleaned.csv")
filtered_df = signals[signals['Signal Id'] == "1885162"]
signals.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(19.20, 10.80), dpi=500)

# Assuming you have your data stored in a DataFrame called 'df'
plot = sns.violinplot(data=signals, y='SignalGrowth', inner='quartile',)

figure = plot.get_figure() 
# Set labels and title
plt.xlabel('')
plt.ylabel('Signal Growth %')
plt.title('Distribution of Signal Growth %')
# Display the plot
plt.savefig('Signal Growth Violin Dis Full.png', dpi=500, bbox_inches='tight')
```

Remove growth that above 1000

```python
df = signals
rows_to_remove = df[df['SignalGrowth'] > 1000].index
# Remove the rows from the DataFrame
df.drop(rows_to_remove, inplace=True)
# Print the updated DataFrame
print(df)
```

New Violin Plot

```python

plt.figure(figsize=(19.20, 10.80), dpi=100)
# Assuming you have your data stored in a DataFrame called 'df'
sns.violinplot(data=df, y='SignalGrowth', inner='quartile', bw=0.1)

# Set labels and title
plt.xlabel('')
plt.ylabel('Signal Growth %')
plt.title('Distribution of Signal Growth %')
plt.savefig('Signal Growth Violin Dis Remove above 1000.png', dpi=300, bbox_inches='tight')
# Display the plot
plt.show()
```
